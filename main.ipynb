## Начало

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
from scipy.io.arff import loadarff 

raw_data = loadarff("Scenario A1-ARFF/TimeBasedFeatures-Dataset-15s-VPN.arff")


# raw_data
df = pd.DataFrame(raw_data[0])
df.head(5)
```

```python
df.describe()
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier


y = df['class1'] # сохраняем метки классов в виде значений s-1, b-0

y = y.replace([b'Non-VPN'], 0)
y = y.replace([b'VPN'], 1)

y
# sum = 0
# count = 0
# for i in y:
#     if i == b'Non-VPN':
#         y.loc[count] = 0
#     else:
#         y.loc[count] = 1
#         sum = sum + 1
#     count = count + 1
# print(sum) # Кол-во элементов, которые принимают единицу

df.drop("class1",axis=1, inplace=True) # удаляем метки классов из датафрейма

x_train, x_test, y_train, y_test = train_test_split(df, y, test_size=0.2, random_state=42) # рабщиваем на тестовую и обучающую выборку. Берем 20% на тест.
# 42 отвечает за то, чтобы результат не менялся при перезапуске модели(данное значение рекомендуется в документации)

model = RandomForestClassifier(n_estimators=100) # берем 100 деревьев для обучения, глубину не ограничиваем
model.fit(x_train, y_train) # обучаем модель 

importances = model.feature_importances_

final_df = pd.DataFrame({"Features" : pd.DataFrame(x_train).columns, "Importances" : importances})
plot_df = pd.DataFrame({"Features" : pd.DataFrame(x_train).columns, "Importances" : importances}, index=pd.DataFrame(x_train).columns)
final_df.set_index('Importances')
final_df = final_df.sort_values('Importances', ascending=False)
plot_df = plot_df.sort_values('Importances', ascending=False)
final_df
```

```python
plot_df.plot.bar(color = 'teal')
```

```python
from sklearn.metrics import accuracy_score 
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.model_selection import cross_validate

predict = model.predict(x_test)
predict_proba = model.predict_proba(x_test)
cv_result = cross_validate(model, df, y, cv = 10)

print(accuracy_score(y_test, predict)) # точность accuracy 
print(recall_score(y_test, predict)) # полнота recall
print(f1_score(y_test, predict)) # f-мера/f1-мера, что одно и то же f-score
print(f"Точность при 10 выборках: {cv_result['test_score']}\n" 
      f"Средняя точность: {cv_result['test_score'].mean()}") # тестирование кросс-валидацией (берем среднее значение)
# print(predict) 
```

```python
from sklearn.metrics import roc_curve


fprs, tprs, thr = roc_curve(y_test, predict_proba[:, 1])

plt.plot(fprs, tprs, marker='*')
plt.ylabel('TruePositiveRate') # Кол-во найденных положительный значений, от всех положительных 
plt.xlabel('FalsePositiveRate') # Кол-во найденных отрицательных значений от всех, отрицательных
plt.title('ROC-curve') # Чем выше кривая, тем лучше алгоритм
```

```python
from sklearn.metrics import auc

print(f"Площадь под графиком: {auc(fprs,tprs)}")
```
